
# **Naive rule**

```python
#Count the number of two classes
df['diagnosis'].value_counts()
```

```python
#percentage of two classes
class0=357/(357+212)
class1=212/(357+212)
print('The percentage of class 0 is ', class0)
print('The percentage of class 1 is ', class1)
```

```python
#plot the naive rule by value
plt.bar(['0','1'],[357,212])
plt.xlabel('diagnosis')
plt.ylabel('count')
plt.title('Plot the naive rule by value')
```

```python
#plot the naive rule by percentage
plt.bar(['0','1'],[class0,class1])
plt.xlabel('diagnosis')
plt.ylabel('percentage')
plt.title('Plot the naive rule by percentage')
```
